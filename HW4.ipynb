```python
# Consider Picking 10 random songs from each genre when evaluating different song pools

random_songs = df.sample(1000, random_state = 10)
random_songs
```

```python
random_songs = random_songs.iloc[:,4:]
random_songs
```

```python
random_songs.index
```

```python
random_songs['explicit'] = random_songs['explicit'] * 1
random_songs
```

```python
genre_df = pd.get_dummies(random_songs['track_genre'])
```

```python
# Create new column with similarity index
new_df = pd.DataFrame(columns=['song1', 'song2', 'distance'])

# Iterate through each row
for index, row in random_songs.iterrows():
    
    # Get the song name
    song1 = row['track_name']
    
    # Get the song features
    features1 = row[2:]
    
    # Iterate through each row again
    for index2, row2 in random_songs.iterrows():
        
        # Get the song name
        song2 = row2['track_name']
        
        # Get the song features
        features2 = row2[2:-1]
        
        # Calculate the distance
        distance = math.sqrt(sum((features1 - features2)**2))
        
        # Append song1, song2, and distance to the new dataframe
        new_df = new_df.append({'song1': song1, 'song2': song2, 'distance': distance}, ignore_index=True)
```
