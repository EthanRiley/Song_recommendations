```python
import math
import pandas as pd
#silence warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('spotify.csv')
```

```python
# Remove first column
data = data.iloc[:,1:]
```

```python
data
```

```python
# Get a randome sample of 1000 songs
data = data.sample(n=1000)
```

```python
# Reduce data frame to just the columns popularity, duration_ms, explicit, danceability, energy, key, loudness, mode, speechiness, acousticness, instrumentalness, liveness, valence, tempo, time_signature, track_genre

# Create a new dataframe
df2 = data[['track_id', 'track_name', 'popularity', 'duration_ms', 'explicit', 'danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo', 'time_signature', 'track_genre']]
```

```python
data[['track_name', 'popularity', 'duration_ms', 'explicit', 'danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo', 'time_signature', 'track_genre']]
```

```python
# Create a column for each genre with value 1 if it is that genre and 0 otherwise
for genre in data['track_genre'].unique():
    df2[genre] = data['track_genre'].apply(lambda x: 1 if x == genre else 0)
```

```python
df2
```

```python
# Remove track_genre column
df2 = df2.drop(columns=['track_genre'])
```

```python
# Convert explicit column to 1 if True, 0 if False
df2['explicit'] = df2['explicit'].apply(lambda x: 1 if x == True else 0)
```

```python
df2
```

```python
# Use min-max scaling to scale data
# Iterate through each column
for column in df2.columns:
    if column == 'track_name' or column == 'track_id':
        continue
    # Get the min and max values
    min_val = df2[column].min()
    max_val = df2[column].max()
    # Scale the values
    df2[column] = df2[column].apply(lambda x: (x - min_val) / (max_val - min_val))

df2
```

```python
# Create a new dataframe
connection_df = pd.DataFrame(columns=['song1', 'song2', 'distance'])

# Create a version of df 2 that is the first 200 rows

# Iterate through each row
# Seperate first 100 rows of data
iteration = 1

for index, row in df2.iterrows():
    # Get the song name
    song1 = row['track_name']
    # Get the song features
    features1 = row[2:]
    # Iterate through each row again
    for index2, row2 in df2.iterrows():
        # Get the song name
        song2 = row2['track_name']
        # Get the song features
        features2 = row2[2:]
        # Calculate the distance
        distance = math.sqrt(sum((features1 - features2)**2))
        # Append song1, song2, and distance to the new dataframe
        connection_df = connection_df.append({'song1': song1, 'song2': song2, 'distance': distance}, ignore_index=True)
    # Print the iteration if it is a multiple of 10
    iteration += 1
    if iteration % 50 == 0:
        print(iteration)

    
```

```python
# Save connection_df to a csv file
connection_df.to_csv('connection_df.csv')
```
